The **information** variable encodes the reaction being modelled, and the **FullAnalysisRun** will print:
- A full simulation, plotting a table with; all the molecular species, their initial concentrations, and the final concentrations.
- A table of the rate values used/
- All the ODE's of the system
- The sensitivity and normalised sensitivity Heatmaps
- The rate parameter estimation

The variable **information** is a list of lists. Where each list describes the variable, its production process, and its miRNA regulation. 

**Examples:**

- So for a simulation of RFP production and GFP production:  [ [ "R" , "full" , " " ] ,   [ "G" , "full" , " " ] ]

- For a simulation of RFP production, GFP production and 3' miRNA production: [ [ "R" , "full" , " " ] ,   [ "G" , "full" , " " ],   [ "3" , "mRNA" , " " ] ]

- For a simulation of RFP production, GFP production and 3' miRNA production. Whereby RFP is being regulated by 3' miRNA:  [ [ "R" , "full" , "3" ] ,   [ "G" , "full" , " " ],   [ "3" , "mRNA" , " " ] ]

- For a simulation of RFP production, GFP production and 5' miRNA production. Whereby RFP is being regulated by 5' miRNA: [ [ "R" , "full" , "5" ] ,   [ "G" , "full" , " " ],   [ "5" , "mRNA" , " " ] ]
